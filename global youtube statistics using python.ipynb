DATA Visualizaton

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
ax=sns.barplot(y='subscribers',x='Youtuber', data=df,order=df.sort_values('subscribers',ascending=False).Youtuber.iloc[:10],palette='magma')
plt.xticks(rotation=60)
ax.set_yticklabels([f'{int(label/1e6)}M' for label in ax.get_yticks()])
plt.title('Top 10 Youtubers in terms of no. of Subscribers')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
ax=sns.barplot(y='video views',x='Youtuber',data=df,order=df.sort_values('video views',ascending=False).Youtuber.iloc[:10],palette='magma')
plt.xticks(rotation=90)
ax.set_yticklabels([f'{int(label/1e9)}B' for label in ax.get_yticks()])
plt.title('Top 10 Youtubers in terms of Views(in billions)')
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(16,6))
sns.countplot(x='category',data=df,order=df['category'].value_counts(ascending=False).index,palette='magma',ax=ax[0])
sns.countplot(x='Country',data=df,order=df['Country'].value_counts(ascending=False).index,palette='magma',ax=ax[1])
ax[0].set_xticklabels(list(df['category'].value_counts(ascending=False).index),rotation=90)
ax[1].set_xticklabels(list(df['Country'].value_counts(ascending=False).index),rotation=90)
ax[0].set_title('No. of Youtube channels from each category')
ax[1].set_title('No. of youtube channels from each country')
```

```python
categories_counts = df['channel_type'].value_counts()

plt.figure(figsize=(8, 6))

plt.pie(categories_counts.head(10), labels=None, autopct='%1.1f%%', startangle=140)

plt.legend(categories_counts.head(10).index, loc='upper right')
plt.title('Top 10 YouTube Channel Types', fontsize=14)
plt.axis('equal') 

plt.show()
```
